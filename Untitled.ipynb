```python
import pandas as pd
import numpy as np
from env import user,host,password

def get_connection(db, username=user, host=host, password=password):
    '''
    Creates a connection URL
    '''
    return f'mysql+pymysql://{username}:{password}@{host}/{db}'
    
def new_salese_data():
    '''
    Returns tsa sales data into a dataframe
    '''
    sql_query = '''
Select * from items
LEFT OUTER JOIN sales
    USING (item_id)
LEFT OUTER JOIN stores
    USING (store_id);
 '''
    df = pd.read_sql(sql_query, get_connection('tsa_item_demand'))
    return df
```
